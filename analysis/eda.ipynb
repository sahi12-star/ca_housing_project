# Exploratory Data Analysis (EDA)

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

# Read the housing data

```python
housing = pd.read_csv("data/train/housing_train.csv")
housing.head()
```

# Geographic Data Visualization

```python
# plot housing data
housing.plot(kind="scatter", x = "longitude", y = "latitude", alpha = 0.6, s=housing["population"]/100, figsize=(14,7), c = "median_house_value", cmap=plt.get_cmap("jet"), colorbar= True)
```

```python
plt.savefig("images/geographic_plot.png")
```

```python
housing.to_csv("data/train/housing_train_processed.csv", index=False) 
```
